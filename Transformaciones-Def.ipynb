```python
# FunciÃ³n para extraer el texto de las revisiones de la lista
def extract_review_text(review_list):
    return ' '.join(review.get('review', '') for review in review_list)

# Aplicar la funciÃ³n para extraer el texto de las revisiones
df_sn_review["review_text"] = df_sn_review["reviews"].apply(extract_review_text)
```

```python
#Transformar con NLP las review

#Palabras positivas y negativas
positive_words = {"good", "great", "excellent"}
negative_words = {"bad", "poor", "terrible"}

# Analizar el sentimiento de una revisiÃ³n y devolver la categorÃ­a
def analyze_sentiment(review):
    # TokenizaciÃ³n y limpieza de texto
    words = review.lower().split()
    cleaned_words = [word.strip(".,!?") for word in words]
    
    # Conteo de palabras positivas y negativas
    positive_count = sum(1 for word in cleaned_words if word in positive_words)
    negative_count = sum(1 for word in cleaned_words if word in negative_words)
    
    # Puntaje de sentimiento
    sentiment_score = positive_count - negative_count
    
    # AsignaciÃ³n de categorÃ­as
    if sentiment_score < -1:
        return 0  # Malo
    elif sentiment_score > 1:
        return 2  # Bueno
    else:
        return 1  # Normal

# Aplicar la funciÃ³n de anÃ¡lisis de sentimientos al DataFrame
df_sn_review["sentiment_analysis"] = df_sn_review["review_text"].apply(analyze_sentiment)
```

```python
import re

# FunciÃ³n para extraer el aÃ±o de la columna 'posted'
def extract_year(posted):
    # Utilizar una expresiÃ³n regular para encontrar el aÃ±o en el texto
    year_match = re.search(r'\b\d{4}\b', posted)
    # Si se encuentra el aÃ±o, devolverlo, de lo contrario devolver NA
    if year_match:
        return year_match.group()
    else:
        return pd.NA

# DataFrame final
df_final['year'] = df_final['posted'].apply(extract_year)
```

```python
import pandas as pd

# Cargar el DataFrame desde un archivo CSV
df_item = pd.read_csv("C:/Users/Federico/Desktop/Trabajo Individual Henry/1/PI MLOps - STEAM/item.csv")
df_review = pd.read_csv("C:/Users/Federico/Desktop/Trabajo Individual Henry/1/PI MLOps - STEAM/review.csv")
df_game = pd.read_csv("C:/Users/Federico/Desktop/Trabajo Individual Henry/1/PI MLOps - STEAM/game.csv")
```

```python
df_item.rename(columns={"item_name": "title"}, inplace=True)
df_item.dropna()
```

```python
# Convertir la columna 'fecha' a tipo datetime sin especificar un formato
df_game["release_date"] = pd.to_datetime(df_game["release_date"], errors='coerce')

# Extraer el aÃ±o de la columna 'fecha' y almacenarlo en una nueva columna llamada 'year'
df_game["year"] = df_game["release_date"].dt.year
df_game.dropna().drop(["release_date"], axis=1)
```

```python
df_fun1 = pd.merge(df_game, df_item, on='title', how='inner')
df_fun1
```

```python
import ast

# Convertir las cadenas de texto en listas de Python
df_fun1['genres'] = df_fun1['genres'].apply(ast.literal_eval)

# Ahora puedes usar el mÃ©todo explode
df_fun1_exploded = df_fun1.explode('genres')
```

```python
# Agrupar por gÃ©nero y aÃ±o, y sumar las horas jugadas
df_genre_year_playtime = df_fun1_exploded.groupby(["genres", "year"])["playtime_forever"].sum().reset_index()

# Encontrar el aÃ±o con la mayor cantidad de horas jugadas para cada gÃ©nero
idx = df_genre_year_playtime.groupby("genres")["playtime_forever"].transform("max") == df_genre_year_playtime["playtime_forever"]
df_max_hours_year = df_genre_year_playtime[idx]

# Renombrar las columnas
df_max_hours_year.columns = ["genre", "year", "playtime_forever"]

# Imprimir el DataFrame resultante
df_max_hours_year.drop(["playtime_forever"], axis = 1)
```

```python
# Guardar el DataFrame como un archivo CSV
df_max_hours_year.to_csv("C:/Users/Federico/Desktop/Trabajo Individual Henry/1/PI MLOps - STEAM/PlayTimeGenres.csv", index=False)
```

```python
def PlayTimeGenre(genre):
    # Filtrar el DataFrame por el gÃ©nero especificado
    df_genre = df_max_hours_year[df_max_hours_year['genre'] == genre]
    
    # Verificar si se encontraron datos para el gÃ©nero especificado
    if df_genre.empty:
        return None
    
    # Obtener el aÃ±o correspondiente al gÃ©nero con mayor tiempo jugado
    aÃ±o_max_playtime = df_genre.loc[df_genre['playtime_forever'].idxmax(), 'year']
    
    return aÃ±o_max_playtime
```

```python
def UserForGenre(genre):
    # Suponiendo que tienes el DataFrame 'df' disponible en este contexto
    genre_df = result_df[result_df['genres'] == genre]
    
    # Encontrar el Ã­ndice del usuario que mÃ¡s jugÃ³ para ese gÃ©nero
    max_playtime_index = genre_df['playtime_forever'].idxmax()
    
    # Obtener el usuario y las horas correspondientes
    max_playtime_user = genre_df.loc[max_playtime_index, 'user_id']
    max_playtime_hours = genre_df.loc[max_playtime_index, 'playtime_forever']
    
    return max_playtime_user, max_playtime_hours
```
